## Decision tree models

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
file_train = '/content/sample_data/california_housing_train.csv'
file_test = '/content/sample_data/california_housing_test.csv'
```

```python
df_train = pd.read_csv(file_train)
df_test = pd.read_csv(file_test)
```

```python
df_train.describe()
```

```python
df_train.isnull().sum()
```

```python
df_train['rooms'] = df_train['total_rooms'] / df_train['households']
df_test['rooms'] = df_test['total_rooms'] / df_test['households']
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
train_y = df_train['median_house_value']
train_x = df_train[['housing_median_age', 'rooms', 'median_income']]

test_y = df_test['median_house_value']
test_x = df_test[['housing_median_age', 'rooms', 'median_income']]
```

```python
# specify model
model = DecisionTreeRegressor(random_state = 1)
```

```python
# fit model
model.fit(train_x, train_y)
```

```python
# make validation prediction and calculate MAE
val_prediction = model.predict(test_x)
val_mae = mean_absolute_error(val_prediction, test_y)
print('Validation MAE:', val_mae)
```
